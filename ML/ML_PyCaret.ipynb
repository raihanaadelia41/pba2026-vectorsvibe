```python
from pycaret.classification import *
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/User/OneDrive - Institut Teknologi Sumatera/Documents/mct-nlp-main/mct-nlp-main/module_ML/data/dataset_final_2.csv")

df = df.drop(columns=['Unnamed: 0', 'No.', 'Unnamed: 6'], errors='ignore')

df.head()
```

```python
# 1. BARIS AJAIB: Buang semua kolom sampah, sisakan 2 VIP ini saja!
df_bersih = df[['Teks_Bersih', 'Kategori']]

# 2. Masukkan df_bersih ke dalam PyCaret
clf = setup(
    data=df_bersih,                 # <--- Pastikan pakai data yang sudah difilter
    target="Kategori",
    text_features=["Teks_Bersih"],  # kolom teks
    session_id=123,
    fold=5,
    verbose=False
)
```

```python
best_model = compare_models(
    include=["svm", "nb", "lr"]
)
```

```python
save_model(best_model, "nlp_pipeline_final")
```
